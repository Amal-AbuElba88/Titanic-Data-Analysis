# About the Dataset: 
The Titanic dataset contains details about passengers on the 
Titanic ship. It includes information such as name, age, sex, ticket class, 
fare, port of embarkation, and whether they survived. This dataset is 
helpful for learning how to work with real-world data, especially when 
dealing with missing values and categorical information.

# 1. Loading Data

```python
import pandas as pd
import matplotlib.pyplot as plt
 

# Load Data
df = pd.read_csv("c:/Users/sa/Desktop/python/Second Semester/titanic/train.csv")


# 1. Display Data
print(df.head())
print(df.info())
print(df.describe())
```

# 2. cleaning misssing data

```python
# 2. cleaning misssing data
print("\nThe Number of Missing Data in each columns: ")
print(df.isnull().sum())


# Compensation for the missing values on average
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'] = df['Cabin'].fillna("Unknown")


print("\nAfter cleaning Data:\n", df.isnull().sum())
```

# 3. visual Analysis

```python
# Survival by sex النجاة حسب الجنس
survived_male = df[(df['Sex']=='male') & (df['Survived']==1)].shape[0]
not_survived_male = df[(df['Sex']=='male') & (df['Survived']==0)].shape[0]
survived_female = df[(df['Sex']=='female') & (df['Survived']==1)].shape[0]
not_survived_female = df[(df['Sex']=='female') & (df['Survived']==0)].shape[0]

labels = ['Male Survived', 'Male Not Survived', 'Female Survived', 'Female Not Survived']
counts = [survived_male, not_survived_male, survived_female, not_survived_female]

plt.bar(['Male Survived', 'Male Not Survived'], [survived_male, not_survived_male], color='blue', alpha=0.7)
plt.bar(['Female Survived', 'Female Not Survived'], [survived_female, not_survived_female], color='pink', alpha=0.9)
plt.title("Survival by Sex")
plt.show()
```

```python
# Survival According to the class ticket النجاة حسب الدرجة
survived_by_class = df.groupby(['Pclass', 'Survived'])['PassengerId'].count().unstack()
survived_by_class.plot(kind='bar', stacked=True)
plt.title("Survival by Ticket Class")
plt.xlabel("Ticket Class")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Age distribution
plt.hist(df['Age'], bins=30, color='green', alpha=0.7)
plt.title("Passenger Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Survival by port
survived_by_embarked = df.groupby(['Embarked', 'Survived'])['PassengerId'].count().unstack()
survived_by_embarked.plot(kind='bar', stacked=True)
plt.title("Survival by Boarding Port")
plt.xlabel("Boarding Port")
plt.ylabel("Number of Passengers")
plt.show()
```

# 4. Data collections (by GroupBy)

```python
# Average age by Sex
print("\nAverage Age by Sex:")
print(df.groupby('Sex')['Age'].mean())


# Survival Rate by Gender and Degree
survival_rate = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()
print("\nSurvival Rate by Sex and Ticket Class (using groupby + unstack):")
print(survival_rate)
```

# 5. Additional Analysis:

```python
# The fare according to the degree and gender
fare_stats = df.groupby(["Pclass", "Sex"])["Fare"].agg(["mean", "max", "min", "std"])
print("\nLeverage Stats by Degree and Gender: ")
print(fare_stats)
```

```python
# 6. Show Heatmap
plt.imshow(survival_rate, cmap='YlGnBu', aspect='auto')
plt.colorbar(label='Survival Rate')
plt.xticks(ticks=[0,1,2], labels=survival_rate.columns)
plt.yticks(ticks=[0,1], labels=survival_rate.index)
plt.title("Survival Rate by Sex and Ticket Class")
plt.xlabel("Ticket Class")
plt.ylabel("Sex")
plt.show()
```
